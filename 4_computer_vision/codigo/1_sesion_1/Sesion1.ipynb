## **1. Configuración Inicial**

En esta sección importaremos las librerías necesarias y definiremos algunas funciones auxiliares para visualizar imágenes de forma sencilla.


```python
# Celdas de código
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Función para mostrar imágenes en formato RGB
def mostrar_imagen(img_bgr, titulo="Imagen"):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(img_rgb)
    plt.title(titulo)
    plt.axis('off')
    plt.show()
```

## **4. Extracción y visualización de canales de color**

En esta sección, vamos a separar una imagen (sea la de demostración o la propia del usuario) en sus canales de color (B, G, R) y visualizarlos por separado.


```python
# Escogeremos la imagen "img_demo" por defecto; si quieres usar la del usuario, cambia a img_user
img = img_user.copy()  # o img_user.copy()

# Dividir canales
B, G, R = cv2.split(img)

# Mostrar cada canal (recordar que cada canal es en escala de grises)
mostrar_imagen(cv2.merge([B, B, B]), "Canal B")
mostrar_imagen(cv2.merge([G, G, G]), "Canal G")
mostrar_imagen(cv2.merge([R, R, R]), "Canal R")
```
