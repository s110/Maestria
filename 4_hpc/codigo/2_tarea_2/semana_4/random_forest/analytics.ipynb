```python
# plot fit_time_s_avg vs n_jobs for different n_samples, añadiendo línea ideal para n_samples=800000
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='n_jobs', y='fit_time_s_avg', hue='n_samples', marker='o')
plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel('Number of Jobs (n_jobs)')
plt.ylabel('Average Fit Time (s)')
plt.title('Fit Time vs Number of Jobs for Different Sample Sizes')
plt.legend(title='Number of Samples (n_samples)')

# Línea ideal para n_samples=800000
n_samples_ideal = 800000
df_ideal = df[df['n_samples'] == n_samples_ideal].sort_values('n_jobs')
n_jobs_ideal = df_ideal['n_jobs'].values
fit_time_single_core = df_ideal[df_ideal['n_jobs'] == min(n_jobs_ideal)]['fit_time_s_avg'].values[0]
ideal_times = fit_time_single_core / n_jobs_ideal*4
plt.plot(n_jobs_ideal, ideal_times, 'k--', label='T=1/p')

plt.legend(title='Number of Samples (n_samples)')
plt.tight_layout()
plt.savefig('fit_time_vs_n_jobs_with_ideal.png')
plt.show()
```

```python
# Calcular speedup para cada grupo de n_samples
df['speedup'] = df.groupby('n_samples')['fit_time_s_avg'].transform(lambda x: x.iloc[0] / x)

# Ahora puedes graficar
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='n_jobs', y='speedup', hue='n_samples', marker='o')

for n_samples in df['n_samples'].unique()[-1:]:
    subset = df[df['n_samples'] == n_samples]
    plt.plot(subset['n_jobs'], subset['n_jobs'], linestyle='--', label=f'Teórico')
plt.xlabel('Number of Jobs (n_jobs)')
plt.ylabel('Speedup')
plt.title('Speedup vs Número de jobs')
plt.legend(title='Number of Samples (n_samples)')
plt.tight_layout()
plt.savefig('speedup_vs_n_jobs_theoretical.png')
plt.show()
```

```python
# Calcular speedup y eficiencia
df['speedup'] = df.groupby('n_samples')['fit_time_s_avg'].transform(lambda x: x.iloc[0] / x)
df['efficiency'] = df['speedup'] / df['n_jobs']

# Graficar eficiencia y línea ideal
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='n_jobs', y='efficiency', hue='n_samples', marker='o')
plt.axhline(1, color='k', linestyle='--', label='Ideal Efficiency = 1')
plt.xscale('log', base=2)
plt.xlabel('Number of Jobs (n_jobs)')
plt.ylabel('Efficiency')
plt.title('Eficiencia vs Número de jobs para diferentes tamaños de muestra')
plt.legend(title='Number of Samples (n_samples)')
plt.tight_layout()
plt.savefig('efficiency_vs_n_jobs.png')
plt.show()
```
